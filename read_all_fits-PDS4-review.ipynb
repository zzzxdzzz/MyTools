```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
from pds4_tools import pds4_read

# -------- CHANGE THIS --------
LABEL_DIR = "/Users/zouxd/Meeting/PDS-reviews/2026-Lucy-3iATLAS/lucy.llorri/data_3i_atlas_raw"
# -----------------------------


label_files = sorted(glob.glob(os.path.join(LABEL_DIR, "*.xml")))

def extract_between(text, left, right):
    i = text.find(left)
    if i < 0:
        return None
    i += len(left)
    j = text.find(right, i)
    if j < 0:
        return None
    return text[i:j].strip()

def get_target_fov_count_from_label_xml(label_path):
    # Ensure it's a readable PDS4 label (optional but keeps you aligned with pds4_tools)
    _ = pds4_read(label_path, quiet=True, lazy_load=True)

    # Robust extraction (works whether it is lucy:target_fov_count or target_fov_count)
    with open(label_path, "r", encoding="utf-8") as f:
        xml = f.read()

    # Try lucy namespace form first (matches your example)
    v = extract_between(xml, "<lucy:target_fov_count>", "</lucy:target_fov_count>")
    if v is None:
        # Try without prefix (in case another label serializes differently)
        v = extract_between(xml, "<target_fov_count>", "</target_fov_count>")

    if v is None:
        return None

    try:
        print(v)
        return int(v)
    except Exception:
        return None

x = []
y = []
bad = 0

for i, lbl in enumerate(label_files):
    try:
        tfc = get_target_fov_count_from_label_xml(lbl)
        if tfc is None:
            bad += 1
            continue
        x.append(i)
        y.append(tfc)
    except Exception:
        bad += 1

x = np.array(x, dtype=int)
y = np.array(y, dtype=int)

print(f"Total labels: {len(label_files)}")
print(f"With target_fov_count: {len(y)}")
print(f"Skipped (missing/failed): {bad}")

plt.figure(figsize=(10, 4))
plt.plot(x, y, "o-")
plt.xlabel("File index (sorted by filename)")
plt.ylabel("target_fov_count")
plt.title("Lucy 3I/ATLAS: target_fov_count per label")
plt.grid(True)
plt.tight_layout()
plt.show()
```
