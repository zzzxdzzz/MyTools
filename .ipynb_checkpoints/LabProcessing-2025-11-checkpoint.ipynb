```python
from scipy.io import readsav
import matplotlib.pyplot as plt
import numpy as np

# 1) Read the IDL .sav file
sav = readsav("your_file.sav", python_dict=True)

# 2) Inspect whatâ€™s inside
print(sav.keys())
```

```python
img = sav["image"]          # adjust key name as needed
print(type(img), img.shape)
```

```python
# Example if your .sav contains a structure named 's'
s = sav["s"]
print(s.dtype)              # see the field names
img = s.image               # or s['image'], depending on how it loads
```

```python
plt.imshow(img, origin="lower", cmap="gray")
plt.colorbar()
plt.title("Image from IDL .sav")
plt.show()
```

```python
# Show the first slice / frame
plt.imshow(img[:, :, 0], origin="lower", cmap="gray")
plt.colorbar()
plt.title("First slice")
plt.show()
```

```python
img_float = (img - img.min()) / (img.max() - img.min())
plt.imshow(img_float, origin="lower", cmap="gray")
```
