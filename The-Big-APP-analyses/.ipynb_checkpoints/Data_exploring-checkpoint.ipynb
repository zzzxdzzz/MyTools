### Most of the data comes from 2 US Department of Education projects.

Integrated Postsecondary Education Data system (IPEDS). Under the Higher Education Act, all institutions that participating in Title IV federal student aid programs must complete IPEDS questionnaires.
College Scorecard data. This data comes through:
Federal reporting from institutions, like IPEDS
Data on federal financial aid from the National Student Loan Data System
Tax information from the US Treasury
Some outcomes shown are recorded for a subset of students. Please refer to the language below these outcomes to learn more about what subset of students these outcomes apply to.



```python
import os, re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 可选底图（如果没装 geopandas，会自动降级只画散点，不影响画所有点）
try:
    import geopandas as gpd
    HAVE_GPD = True
except Exception:
    HAVE_GPD = False

import pgeocode

# ===== 你的CSV绝对路径 =====
CSV_PATH = "/Users/zouxd/Documents/GitHub/MyTools/The-Big-APP-analyses/MERGED2007_08_PP.csv"

# ----- 读取并定位列 -----
df = pd.read_csv(CSV_PATH, dtype=str, low_memory=False)

def find_col(cols, candidates):
    s = {c.lower() for c in candidates}
    for c in cols:
        if c.lower() in s:
            return c
    return None

zip_col = find_col(df.columns, ["ZIP", "ZIPCODE", "zip_code", "postal"])
inst_col = find_col(df.columns, ["INSTNM","institution","name","INSTNM_x","INSTNM_y"])

if zip_col is None:
    raise ValueError(f"没有找到 ZIP 列。检测到的列有：{list(df.columns)[:30]} ...")

out = pd.DataFrame({
    "INSTNM": df[inst_col].astype(str) if inst_col else "",
    "ZIP": df[zip_col].astype(str)
})

# ----- 清洗 ZIP，取前5位 -----
def zip5_only(z):
    if pd.isna(z): return np.nan
    m = re.search(r"\d{5}", str(z))
    return m.group(0) if m else np.nan

out["ZIP5"] = out["ZIP"].map(zip5_only)

# ----- 使用 pgeocode 离线编码（对唯一ZIP查询，然后合并回每一行，保证“每行一个点”）-----
nomi = pgeocode.Nominatim("us")
uniq = sorted(set(out["ZIP5"].dropna()))
geo  = nomi.query_postal_code(uniq)
if isinstance(geo, pd.Series):
    geo = geo.to_frame().T
geo = geo.rename(columns={"postal_code":"ZIP5","latitude":"lat","longitude":"lon"})[["ZIP5","lat","lon"]]
df_geo = out.merge(geo, on="ZIP5", how="left")

# 看看有多少行有坐标
valid = df_geo.dropna(subset=["lat","lon"]).copy()
valid["lat"] = pd.to_numeric(valid["lat"], errors="coerce")
valid["lon"] = pd.to_numeric(valid["lon"], errors="coerce")
valid = valid.dropna(subset=["lat","lon"])

print(f"总行数: {len(df_geo):,}；可绘制坐标行数: {len(valid):,}")

# ----- 画图：优先带美国边界；若无法加载边界则画散点 -----
def plot_points_with_optional_us_boundary(points_df, out_png="us_universities_map.png"):
    if points_df.empty:
        raise ValueError("没有可绘制的坐标（lat/lon）。请检查 ZIP 列。")

    # 地图范围（你要求的）
    LON_MIN, LON_MAX = -170, -60
    LAT_MIN, LAT_MAX = 15, 75

    if HAVE_GPD:
        try:
            usa = gpd.read_file(
                "https://naturalearth.s3.amazonaws.com/110m_cultural/ne_110m_admin_0_countries.zip"
            )
            usa = usa[usa["ADMIN"] == "United States of America"].to_crs(epsg=4326)

            gpts = gpd.GeoDataFrame(
                points_df,
                geometry=gpd.points_from_xy(points_df["lon"], points_df["lat"]),
                crs="EPSG:4326"
            )

            ax = usa.plot(figsize=(11,7), edgecolor="black", facecolor="none")
            gpts.plot(ax=ax, markersize=4, alpha=0.5)

            # ✅ 强制设定视窗范围
            ax.set_xlim(LON_MIN, LON_MAX)
            ax.set_ylim(LAT_MIN, LAT_MAX)

            ax.set_title("U.S. Universities by ZIP Location", pad=10)
            ax.set_xlabel("Longitude"); ax.set_ylabel("Latitude")
            plt.tight_layout(); plt.savefig(out_png, dpi=220); plt.show()
            print(f"地图已保存：{os.path.abspath(out_png)}")
            return
        except Exception as e:
            print("加载美国边界失败，改为仅散点；原因：", e)

    # ---- 纯散点 fallback（无 geopandas 也能画）----
    plt.figure(figsize=(11,7))
    plt.scatter(points_df["lon"], points_df["lat"], s=4, alpha=0.5)

    # ✅ 同样强制视窗范围
    plt.xlim(LON_MIN, LON_MAX)
    plt.ylim(LAT_MIN, LAT_MAX)

    plt.title("U.S. Universities by ZIP Location")
    plt.xlabel("Longitude"); plt.ylabel("Latitude")
    plt.grid(True, linewidth=0.3)
    plt.tight_layout(); plt.savefig(out_png, dpi=220); plt.show()
    print(f"地图已保存：{os.path.abspath(out_png)}")

plot_points_with_optional_us_boundary(valid, out_png="us_universities_map.png")

# 可选：把带坐标的数据保存一份，方便排查哪些行缺ZIP/坐标
df_geo.to_csv("universities_with_coords.csv", index=False)
print("已输出清洗数据：", os.path.abspath("universities_with_coords.csv"))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pgeocode
import geopandas as gpd
import re, os

# ===== 文件路径 =====
CSV_PATH = "/Users/zouxd/Documents/GitHub/MyTools/The-Big-APP-analyses/MERGED2007_08_PP.csv"

# ===== 读取全表（保留所有字段）=====
df = pd.read_csv(CSV_PATH, dtype=str, low_memory=False)

# ===== 提取 ZIP, INSTNM, CONTROL, PREDDEG =====
df = df[["INSTNM", "ZIP", "PREDDEG", "CONTROL"]].copy()

# ===== 清洗 ZIP → ZIP5 =====
def zip5_only(z):
    if pd.isna(z): return np.nan
    m = re.search(r"\d{5}", str(z))
    return m.group(0) if m else np.nan

df["ZIP5"] = df["ZIP"].map(zip5_only)

# ===== 使用 pgeocode 离线地理编码 =====
nomi = pgeocode.Nominatim("us")
uniq = sorted(set(df["ZIP5"].dropna()))
geo = nomi.query_postal_code(uniq)
if isinstance(geo, pd.Series):
    geo = geo.to_frame().T
geo = geo.rename(columns={"postal_code": "ZIP5", "latitude": "lat", "longitude": "lon"})[["ZIP5", "lat", "lon"]]

df_geo = df.merge(geo, on="ZIP5", how="left")

# ===== 筛选：只要四年制本科 PREDDEG = 3 =====
df_geo["PREDDEG"] = pd.to_numeric(df_geo["PREDDEG"], errors="coerce")
df_geo = df_geo[df_geo["PREDDEG"] == 3]

# ===== 转换 CONTROL 为颜色 =====
df_geo["CONTROL"] = pd.to_numeric(df_geo["CONTROL"], errors="coerce")
color_map = {1: "red", 2: "blue", 3: "green"}
df_geo["COLOR"] = df_geo["CONTROL"].map(color_map)

# ===== 去掉没有坐标的行 =====
plot_df = df_geo.dropna(subset=["lat", "lon"]).copy()
plot_df["lat"] = pd.to_numeric(plot_df["lat"], errors="coerce")
plot_df["lon"] = pd.to_numeric(plot_df["lon"], errors="coerce")
plot_df = plot_df.dropna(subset=["lat", "lon"])

print(f"最终可视化学校数: {len(plot_df):,}")

# ===== 加美国边界，按 CONTROL 着色 =====
usa = gpd.read_file(
    "https://naturalearth.s3.amazonaws.com/110m_cultural/ne_110m_admin_0_countries.zip"
)
usa = usa[usa["ADMIN"] == "United States of America"].to_crs(epsg=4326)

gpts = gpd.GeoDataFrame(
    plot_df,
    geometry=gpd.points_from_xy(plot_df["lon"], plot_df["lat"]),
    crs="EPSG:4326"
)

fig, ax = plt.subplots(figsize=(12,8))
usa.plot(ax=ax, edgecolor="black", facecolor="none")
gpts.plot(ax=ax, color=plot_df["COLOR"], markersize=5, alpha=0.6)

# ===== 强制限定地图范围 =====
ax.set_xlim(-180, -50)
ax.set_ylim(10, 85)

# ===== 图例 =====
import matplotlib.patches as mpatches
legend_handles = [
    mpatches.Patch(color="red", label="Public (CONTROL=1)"),
    mpatches.Patch(color="blue", label="Private Non-Profit (CONTROL=2)"),
    mpatches.Patch(color="green", label="Private For-Profit (CONTROL=3)")
]
ax.legend(handles=legend_handles, loc="lower left")

ax.set_title("U.S. Four-Year Universities by Control Type (ZIP Geocoded)")
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")

plt.tight_layout()
plt.show()
```

```python
pd.set_option('display.max_rows', None)
df = pd.read_csv(CSV_PATH, dtype=str, low_memory=False)


df_rank = df.copy()   # 从你地理编码后的表继续（包含 PREDDEG / CONTROL）

# 保留四年制本科
df_rank["PREDDEG"] = pd.to_numeric(df_rank["PREDDEG"], errors="coerce")
df_rank = df_rank[df_rank["PREDDEG"] == 3]

# 转换录取率 & SAT 字段为数值
df_rank["SAT_AVG_ALL"] = pd.to_numeric(df_rank["SAT_AVG_ALL"], errors="coerce")
df_rank["ADM_RATE"] = pd.to_numeric(df_rank["ADM_RATE"], errors="coerce")

# 去掉缺失 SAT 或录取率的学校
df_rank = df_rank.dropna(subset=["SAT_AVG_ALL", "ADM_RATE"])

# 计算综合评分：高 SAT + 低录取率 = 高 score
df_rank["score"] = df_rank["SAT_AVG_ALL"] - 1000 * df_rank["ADM_RATE"]

# 排序 & 取前100
top100 = df_rank.sort_values("score", ascending=False).head(100)

print(f"筛选到的前100所学校（已按 SAT 强 + 录取难 排序）：{len(top100)} 所")
top100_display = top100[["INSTNM", "SAT_AVG_ALL", "ADM_RATE", "score"]]
top100_display
```

```python
# 复制 top100，避免修改原表
top100_programs = top100.copy()

# 转换为数值(避免某些值为字符串)
for col in ["CIP42BACHL", "CIP11BACHL", "CIP27BACHL"]:
    top100_programs[col] = pd.to_numeric(top100_programs[col], errors="coerce")

# 生成布尔标记列
top100_programs["HAS_PSYCH"] = top100_programs["CIP42BACHL"] > 0
top100_programs["HAS_CS"]    = top100_programs["CIP11BACHL"] > 0
top100_programs["HAS_MATH"]  = top100_programs["CIP27BACHL"] > 0

# 查看结果（只显示关键列）
top100_programs_result = top100_programs[
    ["INSTNM", "CITY", "STABBR", "SAT_AVG_ALL", "ADM_RATE", "HAS_PSYCH", "HAS_CS", "HAS_MATH"]
]

display(top100_programs_result)
```
